To create a plot using matplotlib, it is normal to create a new figure with the `figure` command. This will open up a new matplotlib window. Use the `figsize` argument to set the size of the figure in inches. The code below creates a simple (and very ugly) plot.

```python
plt.figure()
plt.plot(diamonds.carat,diamonds.price)
```

The default plots in Python are line plots rather than scatter plots. To create a scatter plot we can include a *linestyle* argument; `'.'`, `'o'`, `'-.'`, etc.. We can specify the colour in a similar way; `'r'` (red), `'b'` (blue), `'g'` (green), etc. The below code changes the markers to red dots. 

```python
plt.figure()
plt.plot(diamonds.carat,diamonds.price,'r.')
```

The colour, linestyle and marker can also be specified as separate named arguments, as bellow. Also, we can create a scatter/dot plot by default using the `scatter` function.

```python
plt.figure()
plt.plot(diamonds.carat,diamonds.price,color='r',linestyle='None',marker='.')
```

```python
plt.figure()
plt.scatter(diamonds.carat,diamonds.price,color='g',marker='*')
```

If you want to combine plots onto a single panel, simply add extra plot commands into the code cell.

```python
plt.figure()
plt.scatter(diamonds.carat,diamonds.price,color='g',marker='*')
plt.plot(diamonds.carat,2*diamonds.price,color='r',linestyle='None',marker='.')
```

# Plotting extras
A plot is a graphically representation of data. A good plot is interpretable without a caption or description. To be interpretable a plot must have axis labels and a legend if more than one dataset is depicted. A title is also advised. 

When plotting multiple datasets, it is good practice to include `label` as an input argument in the `plot`/`scatter`function. Then when we run the `legend` function, these labels are used to create a legend.

```python
plt.figure()
plt.scatter(diamonds.carat,diamonds.price,color='g',marker='*',label='Data')
plt.plot(diamonds.carat,2*diamonds.price,color='r',linestyle='None',marker='.',label='Data x2')
plt.ylabel('Price',fontsize=12)
plt.xlabel('Carat',fontsize=12)
plt.title('Scatter plot of price vs carat for Diamonds data',fontsize=14)
plt.legend()
```

Axis limits can be easily change using the `axis` command. The axis limits are specified as a list, $[x_{min},x_{max},y_{min},y_{max}]$, and given as the function argument. The commands `xlim` and `ylim` can be used to change the $x$ or $y$ axis limits separately. These commands take two separtae input arguments, the minimum first follwoed by the maximum.

```python
plt.figure()
plt.scatter(diamonds.carat,diamonds.price,color='g',marker='*',label='Data')
plt.plot(diamonds.carat,2*diamonds.price,color='r',linestyle='None',marker='.',label='Data x2')
plt.ylabel('Price',fontsize=12)
plt.xlabel('Carat',fontsize=12)
plt.title('Scatter plot of price vs carat for Diamonds data',fontsize=14)
plt.legend()
plt.axis([0,3,0,30000])
```

If the default tick marks are not to your satisfaction, you can use `xticks` and `yticks` to change both the location of the tick and the label associated with it.


```python
plt.figure()
plt.scatter(diamonds.carat,diamonds.price,color='g',marker='*',label='Data')
plt.plot(diamonds.carat,2*diamonds.price,color='r',linestyle='None',marker='.',label='Data x2')
plt.ylabel('Price',fontsize=12)
plt.xlabel('Carat',fontsize=12)
plt.title('Scatter plot of price vs carat for Diamonds data',fontsize=14)
plt.legend()
plt.axis([0,3,0,30000])
plt.xticks(np.linspace(0,3,7))
plt.yticks([0,10000,20000,30000],['Free!','Cheap','Pricey','Expensive'])
```

**Note:** I'm using the Numpy function `linspace` here to create the range of values. Unlike `range` and `np.arange` the end value is included in the list/array of values, and the third argument is the number of points you want, rather than the spacing between them.

You can also add text and gridlines to you graph, as shown below. 

```python
plt.figure()
plt.scatter(diamonds.carat,diamonds.price,color='g',marker='*',label='Data')
plt.ylabel('Price',fontsize=12)
plt.xlabel('Carat',fontsize=12)
plt.title('Scatter plot of price vs carat for Diamonds data',fontsize=12)
plt.axis([0,6,0,20000])
plt.xticks(np.linspace(0,6,7))
plt.yticks([0,5000,10000,15000,20000],['Free!','Cheap','Pricey','Expensive','Ouch!'])
plt.text(3.3, 13000, 'Very big diamond!')
plt.grid(b=True, which='major',color='0.6', linestyle=':')
```

# Bar charts and histograms
To create a bar chart, we first need to learn about the `value_counts` function/method. When applied to a Series, `value_counts` returns a Series where the indices are the unique values in the Series it was applied to and the values are the number of times each value appearred in that Series.

```python
print(diamonds.color.value_counts())
```

Now, this Series can be used to create a bar chart. The command for creating a bar chart is also `plot`, but the argument `kind` must be specified as 'bar'. The plot properties can be changed in a similar fashion as before.

```python
diamonds.color.value_counts().plot(kind ='bar',color='b',alpha=0.4)
plt.xlabel('Colour')
```

Barcharts can be used for categorical and discrete data, as there is a finite number of possible values. For continuous data, the data must be *binned* to create a histogram. Fortunately, `hist` does the binning for use. The default is to create 10 evenly space bins between the minimum and maximum values in the dataset. You can change this, either by giving a number or a list as the `bin` argument. If a number is given, this number of evenly spaced bins are created. If a list is given, the values in the list are used as the beginning and end of each bin.

```python
from os import stat
from pstats import Stats


diamonds.price.hist(bins=[0,2000,5000,8000,13000,15000,20000],color='r',alpha=0.5,label='Unequal bins')
diamonds.price.hist(bins=20,color='b',alpha=0.5,label='20 evenly spaced bins')
plt.legend()
plt.xlabel('Price')
```

A kernel density estimation (KDE) is a continuous version of a
histogram where the bin-width is set and then *slid* across the x-axis. Below we create a *normalised* histogram, which shows the proportion of entries within each bin, rather than the amount, and compare it to the KDE.

```python
diamonds
```

```python
diamonds.price.hist(bins=20,density=True,color='b',alpha=0.2)
diamonds.price.plot(kind='kde',style='k--',linewidth=2.0)
plt.axis([0,25000,0,0.0004])
```

# Other types of plots
## Boxplots
Boxplots can be very useful for comparing a quantitative variable over
various categorical/ordinal groups. The box represents the lower and upper quartiles,
and the central line the median. The whiskers usually
represent the maximum and minimum. Outliers are given as open circles. Unfortunately, it is quite difficult to change te colours, linestyles, etc.

```python
diamonds.boxplot('price',by='color')
```

## Matrix scatter plot
A matrix scatter plot is useful for discerning relationships between many dimensions. It plots each of the numeric columns of your DataFrame against one another. The argument `diagonal` can be used to specify the type of plot you want along the diagonal, rather than plotting a columns against itself.

```python
pd.plotting.scatter_matrix(diamonds,diagonal='hist',color='g', alpha = 0.4, figsize=(12,12));
```

# Seaborn
Seaborn is another plotting package in Python. It is based on matplotlib, but has great functionalities and creates prettier plots. We will look at a couple of examples, but for a more thorough tutorial on seaborn, see the [seaborn documenation](https://seaborn.pydata.org/tutorial.html). 

**Note:** There are a lot of errors in the tutorials. I think it was written for an older version of the package. I recommend looking at the function help files if you are unsure of the syntax.

Seaborn a handful of different plotting styles; dark, white, whitegrid, darkgrid, etc. We'll use darkgrid for now.

```python
import seaborn as sns
sns.set_style("darkgrid")
```

## Plotting relationships

To plot a relationship between two variables we use `relplot`. Unlike matplotlib, the default is to create a scatter plot. Other functions for plotting 2D relationships are `scatterplot` and `lineplot`. All of these functions are called using the syntax below.

```python
sns.relplot(x="carat", y="price", data=diamonds);
```

A particularly useful feature of seaborn is that it allows you colour the data points based on another column in your DataFrame. With matplot lib we would be forced to separate the data and plot each group separately.

As seaborn is based on matplotlib, we can still use the plot formatting tools from matplotlib.

```python
sns.relplot(x="carat", y="price", data=diamonds, hue='color',style='cut')
plt.axis([0,3,0,20000])
plt.xlabel('Carat',fontsize=14)
plt.ylabel('Price',fontsize=14);
```

Seaborn creates a slightly different form of line plot. Instead of joining up all of the data points, it finds the average value of $y$ for each value of $x$ and plots that. 
To create a line plot, include the argument `kind="line"` in the `relplot` function, or use the function `lineplot`.

By default, the confidence intervals are included as shaded areas. Using the argument `ci`, you can turn this off (`ci=None`) or use a different measurement such as the standard deviation (`ci='sd'`).

The arguments `hue` and `style` can be used here also, to create different lines for different cut, color or clarity values.

```python
sns.relplot(x="carat", y="price", data=diamonds, kind="line", ci='sd')
plt.axis([0,2.5,0,20000])
```

## Plotting distributions

For histograms, and other distribution-type visualiastions, we use `distplot`. The default plot type for `distplot` is a histogram with the KDE overlaid. You can look up the help file to see all of the available arguments (`sns.distplot?`)

```python
sns.distplot(diamonds["price"],hist=True,kde=True,bins=10,norm_hist=True)
```

We can combine a scatter or line plot with the distributions for each of the variables `jointplot`, as shown below.



```python
sns.jointplot(x='carat',y='price',data=diamonds)
```
